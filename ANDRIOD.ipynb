```python
# ==========================
# ðŸ“Š DATA VISUALIZATION CODE
# ==========================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# === 0ï¸âƒ£ LOAD DATASET ===
data = pd.read_csv('/content/drive/MyDrive/AI_LAB/Android_Malware.csv')

# === 1ï¸âƒ£ BASIC INFO ===
print("âœ… Dataset Loaded Successfully!\n")
print("Shape of Dataset:", data.shape)
print("\nColumns:\n", data.columns)
print("\nNull Values:\n", data.isnull().sum())
print("\nDuplicate Rows:", data.duplicated().sum())

# === 2ï¸âƒ£ PIE CHART (for 'label' or 'class' columns) ===
if 'label' in data.columns or 'class' in data.columns:
    col = 'label' if 'label' in data.columns else 'class'
    data[col].value_counts().plot(
        kind='pie',
        autopct='%1.1f%%',
        colors=plt.cm.Paired.colors,
        startangle=90,
        shadow=True
    )
    plt.title(f'Distribution of {col}')
    plt.ylabel('')
    plt.show()
else:
    print("âš ï¸ No 'label' or 'class' column found for pie chart.\n")

# === 3ï¸âƒ£ HISTOGRAM for all numerical columns ===
data.hist(figsize=(12, 10), bins=20, color='skyblue', edgecolor='black')
plt.suptitle('Histogram of Numerical Features', fontsize=14)
plt.show()

# === 4ï¸âƒ£ CORRELATION HEATMAP (only for numeric columns) ===
plt.figure(figsize=(10, 8))
numeric_data = data.select_dtypes(include=['int64', 'float64'])

if not numeric_data.empty:
    sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
    plt.title('Feature Correlation Heatmap (Numeric Only)')
    plt.show()
else:
    print("âš ï¸ No numeric columns found for correlation heatmap.\n")

# === 5ï¸âƒ£ BAR CHART (Top 10 Categories for each object column) ===
for col in data.select_dtypes(include=['object']).columns:
    plt.figure(figsize=(8, 4))
    data[col].value_counts().head(10).plot(kind='bar', color='teal')
    plt.title(f'Top 10 Categories in {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()

print("\nðŸŽ‰ Visualization Completed Successfully!")

```

### ðŸ§  MACHINE LEARNING (ML) MODELS

1. **Linear Regression** â€“ Predicts continuous numerical values.
2. **Logistic Regression** â€“ Used for binary classification (yes/no).
3. **Decision Tree** â€“ Splits data into branches based on feature values.
4. **Random Forest** â€“ An ensemble of decision trees for better accuracy.
5. **Support Vector Machine (SVM)** â€“ Finds the best boundary between classes.
6. **K-Nearest Neighbors (KNN)** â€“ Classifies based on similarity to nearby points.
7. **Naive Bayes** â€“ A probabilistic classifier, great for text classification.
8. **Gradient Boosting** â€“ Sequentially builds models to reduce errors.
9. **XGBoost** â€“ Optimized and fast gradient boosting algorithm.
10. **AdaBoost** â€“ Combines weak learners to form a strong classifier.

---

### âš™ï¸ DEEP LEARNING (DL) MODELS

11. **Artificial Neural Network (ANN)** â€“ Basic building block of deep learning.
12. **Convolutional Neural Network (CNN)** â€“ Best for image recognition and analysis.
13. **Recurrent Neural Network (RNN)** â€“ Designed for sequential data (like text or time series).
14. **Long Short-Term Memory (LSTM)** â€“ A type of RNN that remembers long-term dependencies.
15. **Gated Recurrent Unit (GRU)** â€“ A simpler version of LSTM with similar performance.
16. **Autoencoder** â€“ Learns compressed representations of data.
17. **Generative Adversarial Network (GAN)** â€“ Generates new, synthetic data (like fake images).
18. **Transformer** â€“ Uses attention mechanism, foundation for GPT and BERT models.
